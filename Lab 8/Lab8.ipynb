```python
plt.grid(True)
plt.plot(dataframe.head(125)['index'], dataframe.head(125)['release_year'])
plt.title('INDEX / RELEASE YEAR')
plt.xlabel('Index')
plt.ylabel('Release Year')
# plt.xticks(range(0, max(dataframe.head(125)['release_year']) + 1, 250))
# I didn't want to use xticks, Because it isn't eye catchy
```

```python
plt.grid(True)
plt.scatter(dataframe.head(125)['runtime'], dataframe.head(125)['release_year'])
plt.title('Runtime and RELEASE YEAR')
plt.xlabel('Runtime')
plt.ylabel('Release Year')
```

```python
type_color = {"MOVIE": "red",
               "SHOW": "yellow"
}
for type, type_color in type_color.items():
    type_dataframe = dataframe[dataframe['type'] == type]
    plt.scatter(type_dataframe.head(125)['index'], type_dataframe.head(125)['release_year'], c = type_color, label = type)

plt.grid(True)
plt.title('Runtime and RELEASE YEAR')
plt.xlabel('Runtime')
plt.ylabel('Release Year')
plt.legend()
```

```python
sns.histplot(data = dataframe.head(125), x = 'release_year')
plt.title('Release year of every movie')
plt.xlabel('Year')
plt.ylabel('Count')
```

```python
sns.histplot(data = dataframe.head(125), x = 'release_year')
plt.title('Release year of every movie')
plt.grid(axis = 'y')
plt.grid(axis = 'x')
plt.xlabel('Year')
plt.ylabel('Count')
plt.yticks(range(0,33, 3))
```

```python
sns.pairplot(dataframe.head(125), hue = 'release_year')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(x = 'type', y = 'release_year', data = dataframe.head(125), palette = 'Set1')
plt.title('Box Plot')
plt.xlabel('type')
plt.ylabel('Release_Date')
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(x = 'type', y = 'release_year', data = dataframe.head(125), palette = 'Set1')
sns.swarmplot(x = 'type', y = 'release_year', data = dataframe.head(125), color = '.25')
plt.title('Swarm Plot')
plt.xlabel('type')
plt.ylabel('Release_Date')
```

```python
plt.figure(figsize = (8, 6))
sns.kdeplot(data = dataframe.head(125), x = 'release_year', hue = 'type', fill = True, common_norm = False, alpha = 0.4)
plt.title('KERNEL DENSITY PLOT')
plt.xlabel('release_year')
plt.ylabel('Density')
```

```python
sns.violinplot(x = 'type', y = 'release_year', data = dataframe.head(125))
plt.title('VIOLIN PLOT')
```
